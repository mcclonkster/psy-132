# Psychological Disorders and Treatment – Chapters 14–15

# Understanding Psychological Disorders

## Why This Topic Matters

-   Psychological problems affect real people’s lives.
-   Psychology studies these problems using **evidence**, not
    stereotypes.
-   In this part of the course, we ask:
    -   What counts as a **psychological disorder**?
    -   How do we decide?
    -   How can psychology help?

Set the frame for the unit.

-   Acknowledge that mental health is personally relevant for many
    students.
-   Emphasize two themes:
    -   Understanding, not judging.
    -   Using research evidence, not rumors or media clichés.
-   You do not need to introduce many technical terms yet; this slide is
    about purpose and safety.

## What We Are Studying

-   Not every problem in life is a disorder.
-   We focus on patterns of thoughts, feelings, and behaviors that:
    -   Are very **intense** or **extreme**.
    -   Are **long-lasting**.
    -   Are hard to control and live with.

Connect to the concept of psychopathology.

-   Introduce **psychopathology** as the scientific study of mental,
    emotional, and behavioral disorders.
-   Define a **psychological disorder** in student-friendly terms:
    -   A pattern of thoughts, feelings, or behaviors that causes
        serious distress and/or seriously interferes with daily life.
-   Make the distinction between everyday stress and disorder-level
    patterns based on intensity, duration, and impact.

## How We Decide

-   There is no single “abnormal” test.
-   Psychologists do not just ask “Is this weird?”
-   Instead, they look for several **clues**:
    -   How **unusual** something is.
    -   How much it **breaks social rules**.
    -   How much **distress** it causes.
    -   How much it **interferes** with life.

Introduce the framework for thinking about “abnormal.”

-   Explain that the textbook organizes these clues into four ideas:
    -   **Statistical abnormality**
    -   **Social nonconformity**
    -   **Subjective discomfort**
    -   **Maladaptive behavior**
-   Tell students: “We will translate these labels into questions you
    can actually use.”
-   Do not overload the slide with the technical terms; name them
    verbally here.

## How Unusual Is It?

-   Ask:
    -   Is this pattern very **rare** for most people?
    -   Is it far from what is typical in this culture and age group?
-   Being different is not automatically bad.
-   Extreme patterns can still be important clues.

Tie this clue to the textbook term **statistical abnormality**.

-   Define **statistical abnormality** as behavior that is far from the
    average or typical range in the population.
-   Give neutral examples:
    -   An extremely high level of persistent anxiety compared to most
        people.
    -   A very low need for sleep with no tiredness.
-   Clarify that rare does not equal disordered; some rare traits (for
    example, exceptional talents) are not disorders.

## How Much Does It Break Social Rules?

-   Ask:
    -   Does this behavior strongly **violate social norms**?
    -   Does it cause serious problems with other people?
    -   Or is it simply unusual or unconventional?

Connect to the idea of **social nonconformity**.

-   Define **social nonconformity** as behavior that seriously violates
    social norms or expectations.
-   Give generic examples:
    -   Shouting at strangers in a quiet public place.
    -   Ignoring basic rules of public behavior in ways that cause
        conflict or risk.
-   Emphasize that some nonconformity is healthy; it becomes a concern
    when it is extreme and disruptive.

## How Much Pain and Interference?

-   Ask:
    -   How much **emotional pain** is there?
    -   How much does this pattern **get in the way** of:
        -   Work or school?
        -   Relationships?
        -   Basic self-care?
-   Little pain and little interference usually suggests no disorder.

Bring in **subjective discomfort** and **maladaptive behavior**.

-   **Subjective discomfort**: the person feels ongoing emotional
    distress about their own thoughts, feelings, or actions.
-   **Maladaptive behavior**: the behavior makes it very difficult to
    meet everyday demands (for example, school, work, caregiving).
-   Use brief, generic examples:
    -   Discomfort: feeling constant, overwhelming anxiety that feels
        unbearable.
    -   Maladaptive: missing most classes for weeks due to low mood.
-   Emphasize that distress and interference are often the strongest
    signals that help is needed.

## Putting the Clues Together

-   To think about whether something may be a psychological disorder, we
    look at:
    -   How **unusual** it is.
    -   How much it **breaks norms**.
    -   How much **distress** it causes.
    -   How much it **interferes** with life.
-   These clues are **guides**, not rigid rules.

Map the plain-language clues to the textbook terms.

-   Unusual → **statistical abnormality**.
-   Breaks norms → **social nonconformity**.
-   Distress → **subjective discomfort**.
-   Interference → **maladaptive behavior**.
-   Re-state the idea of a **psychological disorder**:
    -   Typically involves more than one of these clues.
    -   Usually involves ongoing patterns, not just a single reaction to
        a bad day.
-   Invite quick student input: ask for examples of something unusual
    but not maladaptive, or something very interfering but not
    particularly rare.

## How Professionals Classify Disorders

-   Professionals need a shared way to describe patterns of problems.
-   They use a manual that:
    -   Lists mental disorders.
    -   Describes patterns of **symptoms**.
    -   Sets out **criteria** for each diagnosis.

Introduce DSM-5 and diagnosis briefly.

-   Name the manual: **DSM-5** (*Diagnostic and Statistical Manual of
    Mental Disorders, Fifth Edition*).
-   Explain:
    -   **Criteria**: specific symptoms and durations that must be
        present.
    -   **Diagnosis**: a label given when a person’s symptoms match a
        set of criteria.
-   Stress from the notes:
    -   DSM categories are revised as research and societal views
        change.
    -   A diagnosis is a classification tool, not a full description of
        a person.
-   Mention **comorbidity** lightly (having more than one diagnosis at
    the same time) without going into detail yet.

## Many Causes Working Together

-   Most psychological disorders do not have a single cause.
-   We look at:
    -   The **body and brain**.
    -   The **mind and coping skills**.
    -   **Relationships, culture, and stress**.
-   These influences **interact**.

Connect to the **biopsychosocial model** and **diathesis–stress**.

-   **Biopsychosocial model**:
    -   Biological factors: genetics, brain structure, brain chemistry,
        physical illnesses.
    -   Psychological factors: learning history, thought patterns,
        coping skills, personality.
    -   Social and cultural factors: family environment, relationships,
        trauma, socioeconomic conditions, cultural norms.
-   **Diathesis–stress model**:
    -   **Diathesis**: a predisposition or vulnerability (for example,
        genetic risk, early adversity).
    -   **Stress**: life events or ongoing pressures.
    -   Risk increases when vulnerability and stress are both present.
-   Keep it conceptual. Students will revisit these ideas when specific
    disorders and treatments are covered.

# Major Types of Disorders

## Big Picture Map of Disorders

-   We group psychological disorders into broad **categories**.
-   Each category has a typical pattern of thoughts, feelings, and
    behaviors.
-   Focus on the **main pattern and feeling**, not memorizing full
    criteria.

Connect directly to 16e’s structure.

-   Categories to mention:
    -   **Psychotic disorders**
    -   **Mood disorders**
    -   **Anxiety disorders**
    -   **Anxiety-related disorders** (OCD, PTSD, etc.)
    -   **Personality disorders**
    -   **Childhood and neurodevelopmental disorders**
-   Remind students: the goal is to recognize patterns, not to learn
    every symptom line by line.

## Psychotic Disorders

-   **Psychotic disorders** involve a loss of contact with **reality**.
-   People may:
    -   Have **hallucinations** (sensing things that are not there).
    -   Have **delusions** (strong beliefs that are clearly false).
    -   Show **disorganized thinking or speech**.

Draw from the 16e notes on psychotic disorders.

-   Define **hallucinations**: for example, hearing voices when no one
    is speaking.
-   Define **delusions**: for example, firmly believing one is being
    targeted or plotted against without evidence.
-   Introduce **schizophrenia** as a key psychotic disorder:
    -   Often involves hallucinations, delusions, disorganized thought,
        and negative symptoms (such as reduced emotional expression).
-   Emphasize that treatments exist and many people improve with
    support.

## Mood Disorders

-   **Mood disorders** involve major disturbances in **emotion**.
-   Depressive disorders include:
    -   Long-lasting low mood and loss of interest.
    -   Changes in sleep and appetite and low energy.
    -   Feelings of worthlessness or guilt.
-   Bipolar and related disorders include:
    -   Shifts between **depression** and **mania** or hypomania.

Base this on 14.3.

-   Clarify that depressive disorders involve persistent, impairing
    sadness and related symptoms.
-   Define **mania** briefly:
    -   Elevated or irritable mood, increased energy, less sleep, racing
        thoughts, risky behavior.
-   Point out that bipolar disorders can be very serious but are
    treatable with medication and psychotherapy.

## Anxiety Disorders

-   **Anxiety disorders** feature excessive, unrealistic **fear** or
    **anxiety**.
-   Anxiety interferes with daily functioning.
-   Common examples:
    -   Generalized anxiety disorder (chronic, widespread worry).
    -   Panic disorder (repeated panic attacks plus fear of more).
    -   Specific phobias (intense fear of particular objects or
        situations).

Anchor these examples in students’ prior understanding of anxiety.

-   Distinguish normal anxiety from disorder-level anxiety by severity,
    duration, and interference.
-   Clarify each example briefly:
    -   Generalized anxiety disorder: persistent worry about many topics
        most days.
    -   Panic disorder: unexpected panic attacks plus ongoing concern or
        behavior changes related to them.
    -   Phobias: intense fear that leads to avoidance and distress.

## Anxiety-Related Disorders

-   Some disorders share anxiety features but are classified separately.
-   Examples include:
    -   **Obsessive-compulsive disorder (OCD)**.
    -   **Trauma- and stressor-related disorders** such as **PTSD**.

Pull from 14.4’s anxiety-related section.

-   OCD:
    -   **Obsessions**: recurring, unwanted thoughts or urges.
    -   **Compulsions**: repetitive behaviors done to reduce anxiety.
-   PTSD:
    -   Follows trauma.
    -   Core features: intrusive memories, nightmares, avoidance,
        feeling “on guard” or easily startled.
-   Emphasize that, even though classified separately, these still
    involve anxiety and fear as central components.

## Personality Disorders

-   **Personality disorders** involve long-standing patterns of:
    -   Thinking.
    -   Feeling.
    -   Behaving.
-   These patterns are:
    -   **Inflexible** across situations.
    -   Cause significant **distress** or **impairment**.

Link to 14.3’s personality content.

-   Give high-level examples of patterns (without naming all specific
    disorders):
    -   Deep and lasting mistrust of others.
    -   Very unstable relationships and self-image.
    -   Strong need for admiration with little empathy for others.
-   Stress that these patterns appear across many situations and over
    long periods of time.

## Childhood and Neurodevelopmental Disorders

-   These disorders typically **begin in childhood**.
-   They affect:
    -   Attention.
    -   Learning.
    -   Social communication or behavior.
-   Examples include:
    -   **ADHD**.
    -   **Autism spectrum disorders**.
    -   **Specific learning disorders**.

Base this on the notes for childhood and neurodevelopmental disorders.

-   ADHD:
    -   Problems with sustained attention, hyperactivity, or impulsivity
        beyond what is typical.
-   Autism spectrum disorders:
    -   Differences in social communication.
    -   Restricted or repetitive behaviors, with a wide range of
        presentations.
-   Specific learning disorders:
    -   Significant, persistent difficulty in reading, writing, or math,
        despite opportunity and effort.

## Thinking About “Normal” and “Disordered”

-   Everyone feels anxious, sad, or stressed sometimes.
-   A disorder-level problem is usually:
    -   More **intense**.
    -   More **persistent**.
    -   More **interfering**.
-   We ask:
    -   How severe?
    -   How long?
    -   How much does it interfere with life?

This slide consolidates Section 2 of the notes.

-   Encourage students to apply the earlier “four clues” and these
    questions when they think about disorder categories.
-   Emphasize that diagnosis is a structured process guided by criteria;
    it is not about casually labeling people.

# Treatments, Ethics, and Stigma

## Why Treatments Matter

-   Psychological disorders can cause real **suffering**.
-   Many treatments are **effective**.
-   Goals of treatment include:
    -   Reducing symptoms.
    -   Improving daily functioning.
    -   Enhancing quality of life.

Connect to Chapter 15’s aim.

-   Highlight that therapy and medication are not about “fixing”
    personality, but about reducing distress and increasing functioning.
-   Set up the idea of **evidence-based treatment** as treatments tested
    in good studies and shown to help more than no treatment or placebo.

## Psychotherapy Approaches

-   **Psychotherapy** is a professional relationship aimed at helping
    people change thoughts, feelings, and behaviors.
-   Approaches can differ on:
    -   Understanding vs action.
    -   Therapist-led vs client-led.
    -   Open-ended vs time-limited.

Summarize 15.1’s psychotherapy overview.

-   Connect to the “four characteristics” from the notes:
    -   Insight-oriented vs action-oriented.
    -   Directive vs nondirective.
    -   Open-ended vs time-limited.
    -   Individual vs group; in-person vs remote.
-   Explain that the next slides will focus on the main families of
    therapy, not every named variant.

## Action Therapies: Behavior and Cognitive

-   **Behavior therapies** focus on changing **maladaptive behavior**
    using conditioning principles.
-   **Cognitive therapies** focus on changing **distorted thinking** and
    teaching coping skills.
-   **Cognitive-behavioral therapy (CBT)** combines both behavior and
    cognitive methods.

Pull from 15.2 and 15.3.

-   Behavior therapies:
    -   Classical conditioning techniques: aversion therapy, systematic
        desensitization, exposure.
    -   Operant conditioning techniques: reinforcement, shaping, token
        economies, time-out.
-   Cognitive therapies:
    -   Beck’s cognitive therapy: addresses selective perception,
        overgeneralization, all-or-nothing thinking.
    -   REBT (rational-emotive behavior therapy): targets rigid “musts”
        and “shoulds.”
-   Emphasize CBT as a widely used, well-studied example.

## Insight Therapies: Humanistic and Psychodynamic

-   **Humanistic therapies** emphasize:
    -   Personal growth and self-acceptance.
    -   A supportive, accepting, and empathetic relationship.
-   **Psychodynamic therapies** emphasize:
    -   Unconscious conflicts.
    -   The influence of past relationships on current feelings and
        behaviors.

Align with 15.1 and 15.3.

-   Humanistic (for example, client-centered therapy):
    -   Focus on empathy, genuineness, and unconditional positive
        regard.
    -   The client leads exploration; the therapist reflects and
        supports.
-   Psychodynamic:
    -   Based on Freud’s psychoanalysis and later modifications.
    -   May use techniques like exploring recurring themes,
        understanding transference, and examining early experiences.
-   Clarify that both are **insight-oriented**: they focus on
    understanding, not just changing behavior.

## Medical and Biological Treatments

-   **Medications** can reduce symptoms by changing brain chemistry.
-   **Brain stimulation therapies** are used for some severe,
    treatment-resistant conditions.
-   **Psychosurgery** is rare and reserved for extreme cases.

Draw from 15.4.

-   Medications:
    -   Antidepressants, antianxiety drugs, mood stabilizers,
        antipsychotics.
    -   Often combined with psychotherapy.
    -   Require careful monitoring for side effects.
-   Brain stimulation:
    -   Electroconvulsive therapy (ECT) used especially for severe
        depression not helped by other treatments.
-   Psychosurgery:
    -   Now rare, tightly controlled, and considered only after other
        treatments have failed.

## Hospitalization and Community Support

-   **Hospitalization** provides:
    -   A safe, structured setting during crises.
    -   Short-term intensive treatment.
-   **Community mental health programs** provide:
    -   Ongoing support in the community.
    -   Services such as crisis intervention, education, and case
        management.

Continue from 15.4.

-   Explain when hospitalization may be needed:
    -   Serious risk of harm to self or others.
    -   Severe psychotic episodes where safety and stabilization are
        priorities.
-   Community mental health centers:
    -   Aim to prevent or shorten hospital stays.
    -   Provide long-term support for people living with mental illness.

## Ethics in Mental Health Care

-   Ethical practice protects:
    -   **Safety**
    -   **Rights**
    -   **Dignity**
-   Key ideas:
    -   **Confidentiality**
    -   **Informed consent**
    -   Professional **boundaries**
    -   **Cultural responsiveness**

Connect back to Section 4 of your Block 8 notes.

-   Confidentiality:
    -   Information shared in therapy is private, with exceptions
        (serious danger, certain legal requirements).
-   Informed consent:
    -   Clients are told what treatment involves, potential risks and
        benefits, and alternatives.
-   Boundaries:
    -   The therapist–client relationship is professional; therapists
        avoid relationships that could exploit clients.
-   Cultural responsiveness:
    -   Therapists work to understand a client’s cultural background and
        avoid assumptions based on their own norms.

## Skills, Perseverance, and Managing Mental Health

-   Change can be **slow** and uneven.
-   **Perseverance** means continuing to use coping strategies and stay
    engaged with treatment.
-   You can build skills to:
    -   Support your own mental health.
    -   Support others in seeking help.

Based on 14.5 and 15.5.

-   Tie “perseverance” to the idea of grit:
    -   Working toward goals despite obstacles and setbacks.
-   Managing mental health issues:
    -   Using basic counseling skills as a peer: listening, empathy,
        non-judgment.
    -   Knowing when and how to encourage someone to seek professional
        help.
-   Reinforce the message that seeking counseling is a sign of strength,
    not weakness.

## Stigma, Culture, and Evidence

-   **Stigma** can:
    -   Make people less likely to seek or stay in treatment.
    -   Increase shame and isolation.
    -   Contribute to discrimination.
-   Culture and identity shape:
    -   How distress is expressed.
    -   How behavior is interpreted.
    -   How much people trust mental health systems.
-   **Evidence** can help correct myths and support fairer attitudes.

Use Section 5 of your notes.

-   Barriers to getting help:
    -   Cost, limited services, time pressures, language and cultural
        differences, stigma.
-   Clarify what **stigma** is: negative attitudes and stereotypes about
    mental illness.
-   Culture and identity:
    -   Some groups have been over- or underdiagnosed.
    -   Historical mistreatment can create mistrust.
-   Evidence:
    -   Most people with mental illness are not violent.
    -   Many improve with appropriate treatment.
-   Emphasize that accurate, research-based information helps:
    -   Replace fear with understanding.
    -   Encourage supportive responses and help-seeking.
