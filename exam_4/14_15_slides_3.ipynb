# Slides – Psychological Disorders and Treatment

## Understanding Psychological Disorders

### Why This Topic Matters

-   Psychological problems affect real people’s lives in serious ways.
-   Psychology studies these problems using **evidence**, not
    stereotypes or rumors.
-   In this part of the course, we ask what counts as a **psychological
    disorder**, how we decide, and how psychology can help.

Set the frame for the unit.

-   Acknowledge that mental health is personally relevant for many
    students.
-   Emphasize two themes:
    -   Understanding, not judging.
    -   Using research evidence, not media clichés or myths.
-   This slide is about purpose and tone; you do not need to introduce
    many technical terms yet.

### What We Are Studying

-   Not every life problem is a disorder.
-   We focus on patterns of thoughts, feelings, and behaviors that are:
    -   Very **intense** or **extreme**.
    -   **Long-lasting** over time.
    -   Hard to control and live with day to day.

Connect to the concept of psychopathology.

-   Introduce **psychopathology** as the scientific study of mental,
    emotional, and behavioral disorders.
-   Define a **psychological disorder** in student-friendly terms:
    -   A pattern of thoughts, feelings, or behaviors that causes
        serious distress and/or seriously interferes with daily life.
-   Distinguish everyday stress from disorder-level patterns using
    intensity, duration, and impact.

### How We Decide

-   There is no single “abnormal” test.
-   Psychologists do not just ask “Is this weird?” or “Would I do that?”
-   Instead, they look for several clues:
    -   How **unusual** something is.
    -   How much it **breaks social rules**.
    -   How much **distress** it causes.
    -   How much it **interferes** with life.

Introduce the framework for thinking about “abnormal.”

-   Explain that the textbook organizes these clues into four ideas:
    -   **Statistical abnormality**
    -   **Social nonconformity**
    -   **Subjective discomfort**
    -   **Maladaptive behavior**
-   Tell students: “We will translate these labels into questions you
    can actually use.”
-   Keep the slide itself in plain language; name the technical terms
    verbally.

### How Unusual Is It?

-   Ask whether the pattern is **very rare** compared to most people.
-   Ask whether it is clearly far from what is typical in this culture
    and age group.
-   Remember that being different is not automatically bad or
    disordered.

Tie this clue to the textbook term **statistical abnormality**.

-   Define **statistical abnormality** as behavior that is far from the
    average or typical range in the population.
-   Give neutral examples:
    -   An extremely high level of persistent anxiety compared to most
        people.
    -   Unusually low need for sleep without feeling tired.
-   Clarify that being rare does not automatically mean “disorder”; some
    rare traits (for example, exceptional talents) are not pathological.

### How Much Does It Break Social Rules?

-   Ask whether the behavior **seriously violates social norms**.
-   Ask whether it causes repeated conflict or confusion with other
    people.
-   Consider whether it is truly disruptive, not just unconventional.

Connect this clue to **social nonconformity**.

-   Define **social nonconformity** as behavior that strongly violates
    social rules or expectations in a particular culture.
-   Offer generic examples:
    -   Shouting at strangers in a quiet public place.
    -   Ignoring basic rules of public behavior in ways that create risk
        or serious conflict.
-   Emphasize that some nonconformity is healthy; concern grows when
    behavior is extreme and damaging to functioning.

### How Much Pain and Interference?

-   Ask how much **emotional pain** the person reports.
-   Ask how much the pattern **gets in the way** of:
    -   Work or school responsibilities.
    -   Relationships with other people.
    -   Basic self-care and daily tasks.
-   Little pain and little interference usually suggests no disorder.

Bring in **subjective discomfort** and **maladaptive behavior**.

-   **Subjective discomfort**: the person feels ongoing emotional
    distress about their own thoughts, feelings, or actions.
-   **Maladaptive behavior**: the behavior makes it very difficult to
    meet everyday demands, such as attending school or work.
-   Use short examples:
    -   Discomfort: constant, overwhelming anxiety or sadness that feels
        unbearable.
    -   Maladaptive: missing most classes for weeks or neglecting
        hygiene because of low mood.
-   Emphasize that distress and interference are often the strongest
    signals that help is needed.

### Putting the Clues Together

-   To think about whether something may be a psychological disorder, we
    look at:
    -   How **unusual** it is.
    -   How much it **breaks norms**.
    -   How much **distress** it causes.
    -   How much it **interferes** with life.
-   These clues are used as **guides**, not rigid rules.

Map the plain-language clues to the textbook terms.

-   Unusual → **statistical abnormality**.
-   Breaks norms → **social nonconformity**.
-   Distress → **subjective discomfort**.
-   Interference → **maladaptive behavior**.
-   Re-state the idea of a **psychological disorder**:
    -   Typically involves more than one of these clues.
    -   Usually involves ongoing patterns, not just a single reaction to
        a bad day.
-   Invite quick student input: examples of behavior that is unusual but
    not maladaptive, or very interfering but not rare.

### How Professionals Classify Disorders

-   Professionals need a shared way to describe patterns of problems.
-   They use a manual that:
    -   Lists mental disorders.
    -   Describes patterns of **symptoms**.
    -   Sets out **criteria** for each diagnosis.

Introduce DSM-5 and diagnosis briefly.

-   Name the manual: **DSM-5** (*Diagnostic and Statistical Manual of
    Mental Disorders, Fifth Edition*).
-   Explain:
    -   **Criteria**: specific symptoms and durations that must be
        present.
    -   **Diagnosis**: a label given when a person’s symptoms match a
        set of criteria.
-   Stress:
    -   DSM categories are revised as research and societal views
        change.
    -   A diagnosis is a classification tool, not a full description of
        a person.
-   Mention **comorbidity** (having more than one diagnosable disorder
    at the same time), and that DSM sometimes distinguishes **positive
    symptoms** (added experiences) from **negative symptoms** (losses of
    normal functions).

### Many Causes Working Together

-   Most psychological disorders do not have a single cause.
-   We look at:
    -   The **body and brain**.
    -   The **mind and coping skills**.
    -   **Relationships, culture, and stress**.
-   These influences interact over time.

Connect to the **biopsychosocial model** and **diathesis–stress**.

-   **Biopsychosocial model**:
    -   Biological factors: genetics, brain structure, brain chemistry,
        physical illnesses.
    -   Psychological factors: learning history, thought patterns,
        coping skills, personality.
    -   Social and cultural factors: family environment, trauma, life
        stress, socioeconomic conditions, cultural norms.
-   **Diathesis–stress model**:
    -   **Diathesis**: a predisposition or vulnerability (for example,
        genetic risk, early adversity).
    -   **Stress**: life events or ongoing pressures.
    -   Risk increases when vulnerability and stress are both present.
-   Keep it conceptual here. You will revisit these ideas when
    discussing specific disorders and treatments.

## Major Types of Disorders

### Big Picture Map of Disorders

-   Psychological disorders can be grouped into broad **categories**.
-   Each category has a typical pattern of thoughts, feelings, and
    behaviors.
-   Focus on recognizing the **main pattern and feeling**, not
    memorizing every symptom.

Connect to Chapter 14’s structure.

-   Categories to highlight:
    -   **Psychotic disorders**
    -   **Mood disorders**
    -   **Anxiety disorders**
    -   **Anxiety-related disorders** (OCD, PTSD, etc.)
    -   **Personality disorders**
    -   **Childhood and neurodevelopmental disorders**
-   Remind students that more detailed criteria are in DSM and the
    textbook; slides emphasize the big picture.

### Psychotic Disorders

-   **Psychotic disorders** involve a loss of contact with **reality**.
-   People may:
    -   Have **hallucinations** (sensing things that are not there).
    -   Have **delusions** (strong beliefs that are clearly false).
    -   Show **disorganized thinking or speech**.

Draw from 14.2.

-   Define **hallucinations** (for example, hearing voices when no one
    is speaking).
-   Define **delusions** (for example, believing one is being plotted
    against without evidence).
-   Identify **delusional disorders** and **schizophrenia** as key
    psychotic disorders.
-   Note that schizophrenia often includes hallucinations, delusions,
    disorganized thought, and negative symptoms like reduced emotional
    expression.
-   Emphasize that while these disorders can be severe, treatment and
    support can help many people.

### Mood Disorders

-   **Mood disorders** involve major disturbances in **emotion**.
-   Depressive disorders include:
    -   Long-lasting low mood and loss of interest.
    -   Changes in sleep and appetite and low energy.
    -   Feelings of worthlessness or guilt.
-   Bipolar and related disorders include:
    -   Shifts between **depression** and **mania** or hypomania.

Based on 14.3.

-   For depressive disorders, underscore persistence and interference
    with daily life.
-   Define **mania**:
    -   Very elevated or irritable mood, increased energy, less need for
        sleep, racing thoughts, impulsive or risky behavior.
-   Explain that bipolar disorders involve episodes of depression and
    mania or hypomania.
-   Emphasize that mood disorders are treatable and that many people
    benefit from a combination of medication and psychotherapy.

### Anxiety Disorders

-   **Anxiety disorders** feature excessive, unrealistic **fear** or
    **anxiety**.
-   Anxiety interferes with daily functioning.
-   Common examples:
    -   **Generalized anxiety disorder:** chronic, widespread worry
        about many aspects of life.
    -   **Panic disorder:** repeated panic attacks and persistent
        concern about having more.
    -   **Phobias:** intense, irrational fear of specific objects or
        situations.

Align with 14.4.

-   Clarify the difference between everyday anxiety and disorder-level
    anxiety (severity, duration, interference).
-   Briefly unpack each example:
    -   Generalized anxiety disorder: worry that is difficult to control
        and occurs most days for months.
    -   Panic disorder: sudden surges of intense fear plus worry or
        behavior change about future attacks.
    -   Phobias: strong fear that leads to avoidance and distress when
        the object or situation is encountered.

### Anxiety-Related Disorders

-   Some disorders share anxiety features but are classified separately.
-   Examples include:
    -   **Obsessive-compulsive disorder (OCD)**.
    -   **Trauma- and stressor-related disorders** such as **PTSD**.

Continue from 14.4.

-   **OCD**:
    -   **Obsessions**: recurring, unwanted thoughts or images.
    -   **Compulsions**: repetitive behaviors done to reduce anxiety
        from obsessions.
-   **PTSD**:
    -   Follows exposure to trauma.
    -   Common features: intrusive memories, nightmares, avoidance of
        reminders, hyperarousal.
-   Emphasize that anxiety is central to these conditions, even though
    DSM-5 groups them separately.

### Personality Disorders

-   **Personality disorders** involve long-standing patterns of:
    -   Thinking.
    -   Feeling.
    -   Behaving.
-   These patterns are:
    -   **Inflexible** across situations.
    -   Cause significant **distress** or **impairment**.

Based on 14.3’s personality section.

-   Give high-level examples (without naming every specific diagnosis):
    -   Deep mistrust of others.
    -   Chronic instability in relationships and self-image.
    -   Strong need for admiration with little empathy.
-   Stress that these patterns are persistent and show up across many
    different situations in a person’s life.

### Childhood and Neurodevelopmental Disorders

-   These disorders typically **begin in childhood**.
-   They affect:
    -   Attention.
    -   Learning.
    -   Social communication or behavior.
-   Examples include:
    -   **ADHD**.
    -   **Autism spectrum disorders**.
    -   **Specific learning disorders**.

Summarize from neurodevelopmental content.

-   **ADHD**:
    -   Difficulty sustaining attention, hyperactivity, and/or
        impulsivity beyond what is typical.
-   **Autism spectrum disorders**:
    -   Differences in social communication.
    -   Restricted or repetitive behaviors with varied severity.
-   **Specific learning disorders**:
    -   Persistent problems with reading (dyslexia), writing, or math
        that are not due to lack of opportunity.

### Thinking About “Normal” and “Disordered”

-   Everyone feels anxious, sad, or stressed sometimes.
-   A disorder-level problem is usually:
    -   More **intense**.
    -   More **persistent**.
    -   More **interfering**.
-   We ask:
    -   How severe?
    -   How long?
    -   How much does it interfere with life?

Consolidate the “Normal vs disordered” section.

-   Encourage students to apply:
    -   The four abnormality clues.
    -   These three practical questions.
-   Emphasize that the goal is careful thinking, not quick labeling.
-   Make clear that diagnosis is done by trained professionals using
    structured criteria.

## Treatments, Ethics, and Stigma

### Why Treatments Matter

-   Psychological disorders can cause real **suffering** and disruption.
-   Many treatments are **effective** when properly used.
-   Treatment aims to:
    -   Reduce symptoms.
    -   Improve daily functioning.
    -   Enhance quality of life.

Connect to Chapter 15.

-   Highlight that treatment is not about “fixing” someone’s identity.
-   Introduce the idea of **evidence-based treatment**:
    -   Treatments tested in well-designed studies.
    -   Shown to help more than no treatment or placebo on average.

### Psychotherapy Approaches

-   **Psychotherapy** is a professional relationship aimed at helping
    people change thoughts, feelings, and behaviors.
-   Approaches can differ in:
    -   How much they focus on understanding vs action.
    -   How much the therapist directs vs lets the client lead.
    -   How long treatment is planned to last.

Summarize 15.1.

-   Relate these bullets to:
    -   Insight-oriented vs action-oriented.
    -   Directive vs nondirective.
    -   Open-ended vs time-limited.
-   Mention that therapy can also be individual, group, couples/family,
    and can be in-person or remote.

### Action Therapies: Behavior and Cognitive

-   **Behavior therapies** focus on changing **maladaptive behavior**
    using conditioning.
-   **Cognitive therapies** focus on changing **distorted thinking** and
    teaching coping skills.
-   **Cognitive-behavioral therapy (CBT)** combines behavior and
    cognitive methods.

Draw from 15.2 and 15.3.

-   Behavior therapies:
    -   Classical conditioning techniques: aversion therapy, systematic
        desensitization, exposure.
    -   Operant conditioning techniques: reinforcement, shaping, token
        economies, time-out.
-   Cognitive therapies:
    -   Beck’s cognitive therapy and REBT as examples of approaches
        targeting thought patterns.
-   Emphasize CBT as a common, evidence-based approach that blends both
    strategies.

### Insight Therapies: Humanistic and Psychodynamic

-   **Humanistic therapies** emphasize:
    -   Personal growth and self-acceptance.
    -   A supportive, accepting, and empathetic relationship.
-   **Psychodynamic therapies** emphasize:
    -   Unconscious conflicts.
    -   The influence of past relationships on current feelings and
        behavior.

Based on 15.1 and 15.3.

-   Humanistic (e.g., client-centered therapy):
    -   Focus on empathy, genuineness, and unconditional positive
        regard.
    -   The client leads the conversation; the therapist reflects and
        supports.
-   Psychodynamic:
    -   Evolved from psychoanalysis.
    -   Involves exploring recurring themes and how early experiences
        shape current patterns.

### Medical and Biological Treatments

-   **Medications** can reduce symptoms by changing brain chemistry.
-   **Brain stimulation therapies** are used for some severe,
    treatment-resistant conditions.
-   **Psychosurgery** is rare and reserved for extreme cases.

-   Medications:
    -   Antidepressants, antianxiety drugs, mood stabilizers,

-   Brain stimulation:

### Hospitalization and Community Support

-   **Hospitalization** provides:
    -   A safe, structured setting during crises.

-   **Community mental health centers** provide:
    -   Support in everyday settings.
    -   Services like crisis intervention, education, and case

### Ethics in Mental Health Care

-   Ethical practice protects:
    -   **Safety**
    -   **Rights**

## 14.1 Psychopathology: Classification and Causes

**Psychopathology** is the study of mental, behavioral, and emotional
disorders.

### Four things to consider when deciding whether behavior is abnormal

The textbook identifies four key ideas to think about when deciding if
behavior is abnormal:

1.  **Statistical abnormality**
    -   Behavior is rare or falls far from the average.
    -   Example: very extreme scores on a depression scale compared to
        most people.
2.  **Nonconformity (social nonconformity)**
    -   Behavior seriously violates social norms or expectations.
    -   Example: acting in ways that are grossly inappropriate for the
        situation, such as shouting at strangers in a quiet library.
3.  **Subjective discomfort**
    -   The person feels intense, ongoing emotional pain or distress
        about their own thoughts or behavior.
    -   Example: someone feels tormented by constant, unwanted intrusive
        thoughts.
4.  **Maladaptive behavior**
    -   The behavior seriously interferes with a person’s ability to
        function (work, school, relationships, self-care).
    -   Example: someone who stops going to work, cannot maintain
        hygiene, or cannot manage daily tasks because of their symptoms.

**Key points:**

-   These four ideas are guides, not strict rules.
-   A psychological disorder usually involves more than one of these,
    especially subjective discomfort and maladaptive behavior.

### How disorders are classified and diagnosed

-   Psychological problems are classified using the **Diagnostic and
    Statistical Manual of Mental Disorders, Fifth Edition (DSM-5)**.
-   A **diagnosis** is made by checking which symptoms are present and
    which are absent, and seeing if they match the **criteria** for a
    particular disorder.
-   People can be **comorbid**, meaning they show symptoms that fit
    criteria for more than one disorder at the same time.
-   DSM categories are revised over time as research and social norms
    change; categories can be added, changed, or removed.

The DSM describes symptoms in terms of:

-   **Positive symptoms** – behaviors or experiences that are more than
    normal (for example, hallucinations, delusions).
-   **Negative symptoms** – losses or deficits compared to normal (for
    example, flat affect, lack of speech).

### Causes of mental illness and the biopsychosocial model

The chapter emphasizes the **biopsychosocial model**:

-   **Biological factors:** genetics, brain structure and chemistry,
    physical illness.
-   **Psychological factors:** learning history, thoughts, coping
    skills, personality.
-   **Social and cultural factors:** relationships, family environment,
    culture, life stress, trauma, socioeconomic conditions.

The **diathesis–stress** idea fits here: people may have a
predisposition (diathesis) that, combined with stress, leads to a
disorder.

## 14.2 Psychotic Disorders

Psychotic disorders involve a loss of contact with reality.

-   **Delusional disorders**
    -   Main feature: strong, false beliefs that are not shared by
        others and persist despite evidence to the contrary.
    -   Example: believing you are being plotted against without any
        real proof.
-   **Schizophrenia**
    -   Often involves:
        -   Hallucinations (most often hearing voices).
        -   Delusions (paranoid or grandiose beliefs).
        -   Disorganized thought and speech.
        -   Negative symptoms such as reduced emotional expression or
            lack of motivation.
    -   Typically long-term and can be very disabling, but treatment can
        help many people.

## 14.3 Mood and Personality Disorders

### Mood disorders

Mood disorders involve major disturbances in emotion, such as
long-lasting sadness or extreme swings.

-   **Depressive disorders**
    -   Persistent low mood, loss of interest, changes in sleep and
        appetite, fatigue, feelings of worthlessness, possible thoughts
        of death or suicide.
-   **Bipolar and related disorders**
    -   Shifts between depression and **mania** or **hypomania**:
        -   Elevated mood.
        -   Increased energy.
        -   Decreased need for sleep.
        -   Racing thoughts.
        -   Impulsive or risky behavior.

### Personality disorders

-   Long-standing patterns of thinking, feeling, and behaving that:
    -   Are inflexible across situations.
    -   Cause significant distress or impairment.
-   Examples include enduring patterns of:
    -   Distrust.
    -   Instability in relationships and self-image.
    -   Strong need for admiration with lack of empathy.

## 14.4 Anxiety and Anxiety-Related Disorders

### Anxiety disorders

-   Feature excessive, unrealistic fear or anxiety that interferes with
    functioning.

**Examples:**

-   **Generalized anxiety disorder:** chronic, widespread worry about
    many aspects of life.
-   **Panic disorder:** repeated panic attacks and persistent concern
    about having more.
-   **Phobias:** intense, irrational fear of specific objects or
    situations (for example, heights, animals).

### Anxiety-related disorders

These conditions share anxiety features but are classified separately in

**Examples include:**

-   **Obsessive-compulsive disorder (OCD):** recurring unwanted thoughts
    and repetitive behaviors done to reduce anxiety.
-   **Trauma- and stressor-related disorders** such as **PTSD**:
    -   Intrusive memories.
    -   Nightmares.
    -   Avoidance.
    -   Hyperarousal following trauma.

## 15.1 The Origins and Effectiveness of Psychotherapy

### A brief history of psychotherapy

-   Early roots in **psychoanalysis** (Freud): talking about thoughts,
    dreams, and childhood experiences to uncover unconscious conflicts.
-   Later developments include:
    -   **Behavior therapies** (influenced by learning theory).
    -   **Cognitive therapies** (focused on thought patterns).
    -   **Humanistic therapies** (focused on growth and
        self-acceptance).

### Four characteristics that distinguish psychotherapies

Psychotherapies can be described along these dimensions:

-   **Insight-oriented vs action-oriented**
    -   Insight-oriented: focus on understanding thoughts and feelings
        (for example, psychodynamic, humanistic).
    -   Action-oriented: focus on changing behaviors and thoughts
        directly (for example, behavioral, cognitive-behavioral).
-   **Directive vs nondirective**
    -   Directive: therapist takes an active role in guiding the client,
        giving advice, or assigning tasks.
    -   Nondirective: therapist mainly reflects and supports, letting
        the client lead.
-   **Open-ended vs time-limited**
    -   Open-ended: duration is flexible.
    -   Time-limited: set number of sessions.
-   **Individual vs group; in-person vs remote**
    -   Therapy can be one-on-one, in a group, or with couples/families.
    -   It can be delivered face-to-face or via phone/Internet.

### Core features of effective psychotherapy

The effectiveness of therapy depends on:

-   **Client characteristics**
    -   Motivation, expectations, and resources.
-   **Therapist characteristics**
    -   Warmth, empathy, experience, cultural competence.
-   **Therapeutic alliance**
    -   The quality of the working relationship and mutual trust between
        client and therapist.
-   **Nature of the problem**
    -   Some problems respond better to certain approaches than others.

Overall, research shows that psychotherapy is often more helpful than no
treatment at all across many problems.

------------------------------------------------------------------------

## 15.2 Behavior Therapies

Behavior therapies focus on **changing maladaptive behavior**, often
using principles of conditioning.

### Main elements of behavior therapies

-   Behavior therapists apply **classical** and **operant conditioning**
    to directly change problematic behaviors and thoughts.
-   They emphasize behavior change and skill-building more than deep
    “insight.”

### Therapies based on classical conditioning

-   **Aversion therapy**
    -   Pairs an unwanted behavior (for example, heavy drinking) with
        something unpleasant to reduce that behavior.
-   **Exposure therapies**
    -   Help people face feared situations or cues safely, to reduce
        anxiety over time.
    -   Types include:
        -   **Flooding:** facing the fear all at once.
        -   **Systematic desensitization:** gradual exposure paired with
            relaxation.
        -   **Modeling:** observing others confronting fears.

These techniques aim to weaken learned fear responses.

### Therapies based on operant conditioning

-   Use **reinforcement** and other consequences to reduce undesired
    behaviors and increase desired ones.

**Tools include:**

-   Positive reinforcement: rewarding desirable behaviors.
-   Non-reinforcement and extinction: removing rewards to reduce
    unwanted behaviors.
-   Punishment: adding unpleasant consequences to discourage behavior
    (used carefully).
-   Shaping: reinforcing small steps toward a desired behavior.
-   Stimulus control: managing cues that trigger behaviors.
-   Time-out: removing access to reinforcement after misbehavior.

**Examples:**

-   **Token economies:** people earn tokens for positive behaviors and
    exchange them for rewards.
-   **Intensive behavioral intervention (IBI):** structured behavioral
    programs, often used with children with autism.

## 15.3 Cognitive and Humanistic Therapies

### Cognitive therapies

Cognitive therapies focus on **distorted thinking** and teach coping
skills.

**Key approaches:**

-   **Beck’s cognitive therapy**
    -   Targets patterns like selective perception (only seeing the
        negative), overgeneralization, and all-or-nothing thinking.
    -   Helps clients identify and change these thought patterns.
-   **Cognitive-behavioral therapy (CBT)**
    -   Integrates cognitive and behavioral methods.
    -   Clients learn to challenge distorted thoughts and test new
        behaviors.
-   **Rational-emotive behavior therapy (REBT)**
    -   Trains clients to identify irrational beliefs (“musts,”
        “shoulds”) and replace them with more realistic, flexible
        thoughts.

Overall, cognitive therapies aim to modify the thoughts that fuel
distress and to build more effective coping.

### Humanistic therapies

Humanistic therapies are **insight-oriented**:

-   They help clients gain deeper understanding of their own thoughts,
    emotions, and behaviors.
-   Emphasize personal growth, self-acceptance, and finding meaning.

**Common elements:**

-   Seeing the client as basically good and capable of growth.
-   Providing warmth, empathy, and genuine respect.
-   Letting the client lead the exploration while the therapist offers
    support and reflection.

## 15.4 Medical Therapies

### Main types of medical therapies

-   **Pharmacotherapy (medications)**
    -   Use of drugs (for example, antidepressants, antipsychotics, mood
        stabilizers) to change brain chemistry and reduce symptoms.
-   **Brain stimulation therapies**
    -   Methods like electroconvulsive therapy (ECT), used especially
        for severe depression that has not responded to other
        treatments.
-   **Psychosurgery**
    -   Rare surgical procedures on the brain for severe conditions that
        do not respond to other treatments; much less common now and
        tightly controlled.

### Hospitalization and community mental health programs

-   **Hospitalization**
    -   Provides a protected setting for intensive treatment when there
        is serious risk (for example, suicidal intent, severe
        psychosis).
-   **Community mental health centers**
    -   Aim to prevent or reduce the need for hospitalization.
    -   Offer services like education, consultation, crisis
        intervention, and ongoing support in the community.

## 15.5 Psychology and Your Skill Set: Managing Mental Health Issues

This section emphasizes **skills** for dealing with mental health
problems in everyday life.

### Key ideas include:

-   **Basic counseling skills**
    -   Listening carefully, showing empathy, and avoiding judgment when
        someone shares their struggles.
    -   Knowing how to encourage someone to seek professional help when
        needed.
-   **Getting counseling**
    -   Understanding that seeking help is a sign of strength, not
        weakness.
    -   Knowing steps to find support (talking to a trusted adult,
        contacting a counseling center, exploring community resources).

**The message:**

-   Many psychological problems can be helped.
-   You can play a role in supporting others and in getting help for
    yourself when you need it.
